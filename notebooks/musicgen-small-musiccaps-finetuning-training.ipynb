# Inference

```python
from peft import PeftConfig, PeftModel
from transformers import AutoModelForTextToWaveform, AutoProcessor
from transformers import AutoProcessor, MusicgenForConditionalGeneration
import torch
import time
from IPython.display import Audio

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
repo_id = "./ckpts/MusicGen-Small-MusicCaps-finetuning"

processor = AutoProcessor.from_pretrained(repo_id)
model = MusicgenForConditionalGeneration.from_pretrained(repo_id, torch_dtype=torch.float16)

processor = AutoProcessor.from_pretrained(repo_id)
model.to(device)
inputs = processor(
    text=["flute, traditional music",],
    padding=True,
    return_tensors="pt",
).to(device)
t1 = time.time()
audio_values = model.generate(**inputs, #do_sample=True,
                              guidance_scale=7,
                              # temperature=0.7, # 0-2.0
                              # top_p=0.9, # 0-1
                              # top_k=50, # 50-100
                              max_new_tokens=512)
print(time.time() - t1)
sampling_rate = model.config.audio_encoder.sampling_rate
Audio(audio_values[0].squeeze().cpu().numpy(), rate=sampling_rate)
```
